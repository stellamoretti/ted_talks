```python
def all_ratings_tags(list_of_ratings):
    all_ratings = []
    for l in list_of_ratings:
        for dictionary in l:
            all_ratings.append(dictionary['name'])
    return all_ratings
```

```python
def count_total_ratings(list_of_dicts):
    total = 0
    for d in list_of_dicts:
        total += d['count']
    return total
```

```python
df['count_total_ratings'] = df['ratings_list'].map(count_total_ratings)
```

```python
def beautiful(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Beautiful':
            return dictionary['count']

def persuasive(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Persuasive':
            return dictionary['count']

def jawdropping(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Jaw-dropping':
            return dictionary['count']
        
def ingenious(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Ingenious':
            return dictionary['count']

def ok(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'OK':
            return dictionary['count']

def confusing(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Confusing':
            return dictionary['count']

def funny(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Funny':
            return dictionary['count']

def unconvincing(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Unconvincing':
            return dictionary['count']        
        
def inspiring(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Inspiring':
            return dictionary['count']

def longwinded(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Longwinded':
            return dictionary['count']

def fascinating(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Fascinating':
            return dictionary['count']

def informative(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Informative':
            return dictionary['count']        

def courageous(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Courageous':
            return dictionary['count']

def obnoxious(list_of_dicts):
    for dictionary in list_of_dicts:
        if dictionary['name'] == 'Obnoxious':
            return dictionary['count']
```

```python
df['rated_beautiful'] = df['ratings_list'].map(beautiful) / df['count_total_ratings']
df['rated_persuasive'] = df['ratings_list'].map(persuasive) / df['count_total_ratings']
df['rated_jawdropping'] = df['ratings_list'].map(jawdropping) / df['count_total_ratings']
df['rated_ingenious'] = df['ratings_list'].map(ingenious) / df['count_total_ratings']
df['rated_ok'] = df['ratings_list'].map(ok) / df['count_total_ratings']
df['rated_confusing'] = df['ratings_list'].map(confusing) / df['count_total_ratings']
df['rated_funny'] = df['ratings_list'].map(funny) / df['count_total_ratings']
df['rated_unconvincing'] = df['ratings_list'].map(unconvincing) / df['count_total_ratings']
df['rated_inspiring'] = df['ratings_list'].map(inspiring) / df['count_total_ratings']
df['rated_longwinded'] = df['ratings_list'].map(longwinded) / df['count_total_ratings']
df['rated_fascinating'] = df['ratings_list'].map(fascinating) / df['count_total_ratings']
df['rated_informative'] = df['ratings_list'].map(informative) / df['count_total_ratings']
df['rated_courageous'] = df['ratings_list'].map(courageous) / df['count_total_ratings']
df['rated_obnoxious'] = df['ratings_list'].map(obnoxious) / df['count_total_ratings']
```

# Text Preprocessing

```python
def remove_audience(text):
    text = text.replace("(Laughter)", " ")
    text = text.replace("(Applause)", " ")
    return text
```

```python
df['transcripts_removed_audience'] = df['transcript'].map(remove_audience)
```

```python
def lower(text):
    return text.lower()
```

```python
df['lower'] = df['transcripts_removed_audience'].map(lower)
```

```python
def make_sentences(text):
    return sent_tokenize(text)
```

```python
df['sentences'] = df['lower'].map(make_sentences)
```

```python
def make_words(sentences):
    words = []
    for sentence in sentences:
        words.extend(wordpunct_tokenize(sentence))
    return words
```

```python
df['words'] = df['sentences'].map(make_words)
```

```python
def remove_stopwords(words):
    stopwords = nltk.corpus.stopwords.words('english')
    return [w for w in words if w not in stopwords]
```

```python
df['removed_stopwords'] = df['words'].map(remove_stopwords)
```

```python
def remove_punctuation(words):
    punctuation = ['.', ',', '?', '!', '-', '–', "'", '"', '(', ')', ':', ';', '?"', '!"', '."', ',"']
    words = [w for w in words if w not in punctuation]
    return [w for w in words if w.isalpha()]
```

```python
df['removed_punc'] = df['removed_stopwords'].map(remove_punctuation)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stopwords
```

```python
def recompile(words):
    return ' '.join(words)
```

```python
df['recompiled'] = df['removed_punc'].map(recompile)
```

```python
def stem_words(words):
    stemmer = nltk.stem.porter.PorterStemmer()
    return [stemmer.stem(word) for word in words]
```

```python
df['stemmed_words'] = df['removed_punc'].map(stem_words)
```

```python
df['recompiled_stemmed'] = df['stemmed_words'].map(recompile)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        message = "Topic #%d: " % topic_idx
        message += " ".join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()
```

# Sentiment Analysis

A look at how the sentiment changes throughout the course of selected TED Talks.

```python
def get_sentiment(text):
    polarity, subjectivity = TextBlob(text).sentiment
    return [polarity, subjectivity]

def get_polarity(text):
    polarity, subjectivity = TextBlob(text).sentiment
    return polarity

def get_subjectivity(text):
    polarity, subjectivity = TextBlob(text).sentiment
    return subjectivity
```

polarity: negative vs. positive    (-1.0 => +1.0)
 
subjectivity: objective (fact) vs. subjective (opinion) (+0.0 => +1.0)

```python
def get_polarity_talk(sentences):
    polarity_flow = []
    for sentence in sentences:
        polarity_flow.append(get_polarity(sentence))
    return polarity_flow

def get_subjectivity_talk(sentences):
    subjectivity_flow = []
    for sentence in sentences:
        subjectivity_flow.append(get_subjectivity(sentence))
    return subjectivity_flow
```

```python
df['polarity_flow'] = df['sentences'].map(get_polarity_talk)
```

```python
df['subjectivity_flow'] = df['sentences'].map(get_subjectivity_talk)
```

### Ken Robinson: Do schools kill creativity?

The most viewed TED Talk of all time.

```python
dfKen = df[(df['main_speaker'] == 'Ken Robinson') & (df['film_date'] == 1140825600)]
dfKen
```

```python
KenPolarityFlow = dfKen['polarity_flow'][0]
KenSubjectivityFlow = dfKen['subjectivity_flow'][0]
```

```python
def rolling_average_plot(flow_list, window):
    data = {'raw_flow': flow_list}
    df = pd.DataFrame(data)
    return df.rolling(window=window).mean()
```

```python
plt.figure(figsize=(15, 4))
plt.plot(KenPolarityFlow)
plt.plot(rolling_average_plot(KenPolarityFlow, 3))
plt.title("Sentiment Analysis (Polarity) of 'Ken Robinson: Do schools kill creativity?'")
;
```
